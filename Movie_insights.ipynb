```python
import pandas as pd
import matplotlib.pyplot as plt

def generate_plots(user_id):
#     user_id=1
    # Connect to the SQLite database
#     conn = sqlite3.connect('movieInsights.db')

    # Total number of movies watched grouped by genres for the user
    query_genres = '''
    SELECT genre, COUNT(*) as total_movies
    FROM movie
    WHERE user_id = ?
    GROUP BY genre;
    '''
    genres_df = pd.read_sql_query(query_genres, conn, params=(user_id,))

    # Favourite directors for the user
    query_directors = '''
    SELECT director, COUNT(*) as movie_count
    FROM movie
    WHERE user_id = ?
    GROUP BY director
    ORDER BY movie_count DESC;
    '''
    directors_df = pd.read_sql_query(query_directors, conn, params=(user_id,))

    # Top rated movies in descending order for the user
    query_top_rated = '''
    SELECT movie_name, average_rating
    FROM movie
    WHERE user_id = ?
    ORDER BY average_rating DESC;
    '''
    top_rated_df = pd.read_sql_query(query_top_rated, conn, params=(user_id,))

    # Distribution of movies watched before 2000 and after or equal to 2000
    query_year_distribution = '''
    SELECT (CASE WHEN year < 2000 THEN 'Before 2000' ELSE '2000 and After' END) as period, COUNT(*) as movie_count
    FROM movie
    WHERE user_id = ?
    GROUP BY period;
    '''
    year_distribution_df = pd.read_sql_query(query_year_distribution, conn, params=(user_id,))

    # Close the database connection
#     conn.close()

    # Generate the charts

    # 1. Total number of movies watched grouped by genres
    plt.figure(figsize=(10, 6))
    plt.bar(genres_df['genre'], genres_df['total_movies'])
    plt.title('Total Movies Watched by Genre')
    plt.xlabel('Genre')
    plt.ylabel('Total Movies')
    plt.xticks(rotation=45)
    plt.tight_layout()
#     plt.savefig(f'plots/genres_{user_id}.png')
    plt.show()
    plt.close()

    # 2. Favourite directors
    plt.figure(figsize=(10, 6))
    plt.bar(directors_df['director'], directors_df['movie_count'])
    plt.title('Favourite Directors')
    plt.xlabel('Director')
    plt.ylabel('Number of Movies')
    plt.xticks(rotation=45)
    plt.tight_layout()
#     plt.savefig(f'plots/directors_{user_id}.png')
    plt.show()
    plt.close()

    # 3. Top rated movies
    plt.figure(figsize=(10, 6))
    plt.bar(top_rated_df['movie_name'], top_rated_df['average_rating'])
    plt.title('Top Rated Movies')
    plt.xlabel('Movie Name')
    plt.ylabel('Average Rating')
    plt.xticks(rotation=45)
    plt.tight_layout()
#     plt.savefig(f'plots/top_rated_{user_id}.png')
    plt.show()
    plt.close()

    # 4. Distribution of movies watched by year
    plt.figure(figsize=(10, 6))
    plt.pie(year_distribution_df['movie_count'], labels=year_distribution_df['period'], autopct='%1.1f%%')
    plt.title('Distribution of Movies Watched by Year')
    plt.tight_layout()
#     plt.savefig(f'plots/year_distribution_{user_id}.png')
    plt.show()
    plt.close()
```

```python
generate_plots(1)
```
